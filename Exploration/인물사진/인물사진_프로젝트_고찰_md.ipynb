4. Solution

 - Semnetic segmentaion 과 blur / 심도 조절을 통한 사용자 만족을 가져 가는 방법 이 있을 것같다.

 - 사람이 가지고 있는 각자의 미적 감각은 다르며 , 사용자 경험을 통해서 본인이 가장 만족하는 이미지가 있을 거라고 생각한다. 예를 들어 나는 사진을 찍을 때 인물이 나온다고 하더라도 가능한 배경을 살리는 이미지를 좋아하지만 , 아내의 경우는 인물 사진과 풍경 사진이 극명히 갈리는 타입이다.

 - 그렇다면 사용자가 원하는 객체는 Sementic Segmentation 으로 객체를 선택 하게 하고 ,  그 객체를 중심으로 심도 와 blur 등을 사용자가 조절 하게 만들어 주는 방법 이 있을것 같다.

 - 다른 방법으로는 human feedback 을 통한 Learning 을 한번 더 거침으로서 보편적으로 좋아 할 만한 심도/blur 등의 파라미터들을 학습 시켜서 보편적인 shallow focus 를 만드는 방법도 생각 해 볼수 있을 것 같다.

 - 개인적으로는 누군가 다 만들어 주는 이미지 보다는 사용자 참여를 통해서 개인의 만족감을 더 가져 갈수 있는 플랫폼이 더 좋지 않을까 생각한다. 이를 통한 모델 업데이트도 가능하지 않을까도 생각 해 볼  수 있을 것 같다.
